```python
from e3nn import o3
```

```python
distance_embed_dim = 5

tp1 = o3.FullyConnectedTensorProduct(irreps_in1 = f'{distance_embed_dim}x0e + 1e', irreps_in2 = f'{distance_embed_dim}x0e + 1e', irreps_out = '20x0e + 5x1e')
tp2 = o3.FullyConnectedTensorProduct(irreps_in1 = tp1.irreps_out, irreps_in2 = f'{distance_embed_dim}x0e + 1e', irreps_out = '9x0e + 1x1e')

print(tp2)
tp2.visualize()
```
